```python
## import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
## load the dataset
df = pd.read_csv('emails.csv')
df
```

```python
df.info()
```

```python
df = df.drop(columns=['Email No.'])
```

```python
df
```

```python
## regex , regular expression
df = df.replace(to_replace = r'[^a - zA -Z]', value=float('nan'),regex=True)
df.fillna(0, inplace=True)

##convert into integer type
df = df.astype(int)
```

```python
df.info()
```

```python
# split the data into independent and dependent variable    ##classsification
from sklearn.model_selection import train_test_split
X = df.iloc[:, :-1].values  ##  same as X = df.iloc[:, 0:3002].values
y = df.iloc[:, -1].values
```

```python
## split the data into trainig and testing set
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 32)

```

```python
## train the logistic regression
from sklearn.linear_model import LogisticRegression
model_LR = LogisticRegression()
model_LR.fit(X_train, y_train)
```

```python
## make the prediction
pred_LR = model_LR.predict(X_test) 
pd.DataFrame({'Actual': y_test, 'Prediction': pred_LR})
```

```python
## evaluate the model accuracy
from sklearn.metrics import accuracy_score
accuracy_LR =accuracy_score(y_test, pred_LR)

## confusion metrics
from sklearn.metrics import accuracy_score, confusion_matrix
conf_mat = confusion_matrix(y_test, pred_LR)
print(accuracy_LR)
print(conf_mat)
```

```python
# naive bayes
from sklearn.naive_bayes import GaussianNB
model_NB = GaussianNB()
model_NB.fit(X_train, y_train)
```

```python
#predit the test set
y_pred_NB = model_NB.predict(X_test)
pd.DataFrame({'Actual': y_test, 'Prediction': y_pred_NB})
```

```python
accuracy_NB = accuracy_score(y_test,y_pred_NB)
conf_mat_NB = confusion_matrix(y_test,y_pred_NB)
print(accuracy_NB)
print(conf_mat_NB)
```

```python
## K-Nearest Neigtboors
from sklearn.neighbors import KNeighborsClassifier
Knn_model = KNeighborsClassifier(n_neighbors=5)
##trainin the model
Knn_model.fit(X_train,y_train)

```

```python
## make the prediction
pred_Kn = Knn_model.predict(X_test) 
pd.DataFrame({'Actual': y_test, 'Prediction': pred_Kn})
```

```python
accuracy_knn = accuracy_score(y_test,pred_Kn)
conf_mat_knn = confusion_matrix(y_test,pred_Kn)
print(accuracy_knn)
print(conf_mat_knn)
```

```python
## Desition Tree classifier
from sklearn.tree import DecisionTreeClassifier
Decision_tree_gini = DecisionTreeClassifier(criterion='gini',random_state=100, max_depth=3,min_samples_leaf=5)
Decision_tree_max_depth = DecisionTreeClassifier(max_depth=5)
Decision_tree_entropy = DecisionTreeClassifier(criterion='entropy',random_state=100,max_depth=3,min_samples_leaf=5)

```

```python
##training the models
Decision_tree_gini.fit(X_train,y_train)
Decision_tree_max_depth.fit(X_train,y_train)
Decision_tree_entropy.fit(X_train,y_train)
```

```python
## make predictions
y_pred_gini = Decision_tree_gini.predict(X_test) 
y_pred_max_Dept = Decision_tree_max_depth.predict(X_test) 
y_pred_entropy = Decision_tree_entropy.predict(X_test) 
```

```python
pd.DataFrame({'Actual': y_test, 'Prediction': y_pred_gini})
```

```python
##evaluate the gini
accuracy_gini = accuracy_score(y_test,y_pred_gini)
conf_mat_gini = confusion_matrix(y_test,y_pred_gini)
print(accuracy_gini)
print(conf_mat_gini)
```

```python
pd.DataFrame({'Actual': y_test, 'Prediction': y_pred_max_Dept})
```

```python
##evaluate the max_Dept
accuracy_max_Dept = accuracy_score(y_test,y_pred_max_Dept)
conf_mat_max_Dept = confusion_matrix(y_test,y_pred_max_Dept)
print(accuracy_max_Dept)
print(conf_mat_max_Dept)
```

```python
pd.DataFrame({'Actual': y_test, 'Prediction': y_pred_entropy})
```

```python
##evaluate the entropy
accuracy_entropy = accuracy_score(y_test,y_pred_entropy)
conf_mat_entropy = confusion_matrix(y_test,y_pred_entropy)
print(accuracy_entropy)
print(conf_mat_entropy)
```
